# 선형대수학 1.1 벡터기하와 대수

## (1) 벡터란?

### 벡터(vector)는 한 점 A에서 다른 점 B까지의 이동에 대응하는 유향선분이다.
#### 유향선분이란? 길이와 방향을 모두 가지고 있는 선분이다.

벡터 AB 는 아래와 같이 표현된다.

$$\overrightarrow{AB}$$



```python
import numpy as np
import matplotlib.pyplot as plt

A = [1,2]
B = [5,9]

plt.figure(figsize=(10,10))
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(ticks=np.arange(0, 12, step=1))
plt.yticks(ticks=np.arange(0, 12, step=1))

plt.arrow(A[0],A[1],5-1,9-2, head_width=.2, head_length=.2, color='red')
plt.scatter(A[0],A[1],color='red')
plt.text(A[0]+0.25,A[1],'A', fontsize=20)
plt.text(B[0]+0.25,B[1],'B', fontsize=20)

plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.grid()
plt.show()
```

### 위에 그림에서 A를 시점(꼬리), B를 종점(머리)이라고 한다.

### 많은 벡터중에서 시점이 원점인 벡터를 '위치벡터'라고 한다.

```python
O = [0,0]
B = [3,4]

plt.figure(figsize=(10,10))
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xticks(ticks=np.arange(0, 12, step=1))
plt.yticks(ticks=np.arange(0, 12, step=1))

plt.arrow(0,0,B[0],B[1], head_width=.2, head_length=.2, color='red')
plt.scatter(0,0,color='red')
plt.text(A[0]+0.25,A[1]+0.1,'O',fontsize=20)
plt.text(B[0]+0.25,B[1],'B',fontsize=20)
plt.text(B[0]/2-0.35,B[1]/2,'a',fontsize=20)

plt.axhline(0, color='gray', alpha = 0.3)
plt.axvline(0, color='gray', alpha = 0.3)
plt.title("Vector")
plt.grid()
plt.show()
```

### 일반적으로 벡터는 좌표를 사용하여 나타낸다.<br>위의 그림처럼 B = (3,4)이고 벡터는 아래의 식으로 나타난다.
$$ a = \overrightarrow{OB} = [3,4]$$
